```python
import requests
from bs4 import BeautifulSoup
import lxml
```

```python
url = "https://www.imdb.com/list/ls562791311/?sort=user_rating,desc&st_dt=&mode=detail&page=1"
sayfa= requests.get(url).content
```

```python
soup = BeautifulSoup(sayfa, "lxml")
```

```python
print(soup)
```

```python

top_100 = soup.find("div", class_="lister-list").find_all(class_="lister-item mode-detail")
```

```python
film_sra = 1

for film in top_100:
    ad = film.find("h3", class_="lister-item-header").a.text
    tr = film.find("span", class_="genre").get_text().strip()
    star = film.find("span", class_="ipl-rating-star__rating").text.strip()
    ower = film.find("p", class_="").get_text().strip()
    
    print("rank :", film_sra)
    print("film ad :", ad)
    print("tr :", tr)
    print("aklama :", ower)
    print("vote_average :", star)
    
    print("\n\n")
    film_sra += 1
```

```python
import pandas as pd

# Bo bir DataFrame oluturun
veri = pd.DataFrame(columns=["Rank", "Film Ad", "Tr", "Aklama"])

film_sira = 1

for film in top_100:
    adi = film.find("h3", class_="lister-item-header").a.text
    tur = film.find("span", class_="genre").get_text().strip()
    star = film.find("span", class_="ipl-rating-star__rating").get_text().strip()
    owner = film.find("p", class_="").get_text().strip()
    
    # Yeni bir DataFrame oluturun ve mevcut DataFrame'i gncelleyin
    yeni_veri = pd.DataFrame({"Rank": [film_sira], "Film Ad": [adi], "Tr": [tur], "Aklama": [owner], "vote_average": [star]})
    veri = pd.concat([veri, yeni_veri], ignore_index=True)
    
    film_sira += 1

# Verileri Excel dosyasna kaydedin
veri.to_excel("film_bilgileri.xlsx", index=False)
```
